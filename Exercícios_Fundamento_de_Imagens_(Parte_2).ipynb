## a) Importar as bibliotecas


```python
import numpy as np
import pandas as pd
import cv2 as cv
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from skimage import io

from skimage import data
```

## b) Visualizar e exibir a imagem


```python
print("Imagem 1:\n")
img1 = "https://www.nintendo.com/eu/media/images/10_share_images/portals_3/2x1_SuperMarioHub_image1600w.jpg"
img1 = io.imread(img1)
plt.imshow(img1)
```

```python
print("Imagem 2:\n")
img2 = "https://img.freepik.com/fotos-premium/abstracto-de-lampada-criativa-em-fundo-azul-brilhante-ia-generativa_1259851-72.jpg"
img2 = io.imread(img2)
plt.imshow(img2)
```

```python
print("Imagem 3: \n")
img3 = "https://conteudo.imguol.com.br/c/noticias/1c/2022/05/24/imagem-criada-no-imagen-prototipo-do-google-que-cria-imagens-baseadas-em-texto-neste-caso-um-cachorro-corgi-andando-de-bicicleta-na-times-square-usando-oculos-de-sol-e-chapeu-de-praia-1653397634334_v2_900x506.jpg"
img3 = io.imread(img3)
plt.imshow(img3)
```

## c) Imprimir a matriz da imagem


```python
print("Imagem 1:\n")
print(img1)
print("\nImagem 2:\n")
print(img2)
print("\nImagem 3:\n")
print(img3)
```

## d) Imprimir a altura, largura e canais da imagem


```python
print("Imagem 1:\n")
print(img1.shape)
print("\nImagem 2:\n")
print(img2.shape)
print("\nImagem 3:\n")
print(img3.shape)
```

## e) Gerar e exibir o histograma da imagem


```python
print("Imagem 1:\n")
plt.hist(img1.ravel(),256,[0,256])
plt.show()
print("Imagem 2:\n")
plt.hist(img2.ravel(),256,[0,256])
plt.show()
print("Imagem 3:\n")
plt.hist(img3.ravel(),256,[0,256])
plt.show()
```

## f) Transformar a imagem RGB em uma imagem com tons de cinza


```python
from skimage.color import rgb2gray

img_gray1 = rgb2gray(img1)
img_gray2 = rgb2gray(img2)
img_gray3 = rgb2gray(img3)
print("Imagem 1:\n")
plt.imshow(img_gray1, cmap='gray')
plt.show()
print("\nImagem 2:\n")
plt.imshow(img_gray2, cmap='gray')
plt.show()
print("\nImagem 3:\n")
plt.imshow(img_gray3, cmap='gray')
plt.show()
```

## g) Imprimir a matriz e o histograma da imagem com tons de cinza


```python
print("Imagem 1:\n")
plt.hist(img_gray1.ravel(),256,[0,256])
plt.show()
print("Imagem 2:\n")
plt.hist(img_gray2.ravel(),256,[0,256])
plt.show()
print("Imagem 3:\n")
plt.hist(img_gray3.ravel(),256,[0,256])
plt.show()
```

## h) Transformar a imagem RGB em uma imagem HSV


```python
img_hsv1 = cv.cvtColor(img1, cv.COLOR_RGB2HSV)
plt.imshow(img_hsv1)
plt.show()

img_hsv2 = cv.cvtColor(img2, cv.COLOR_RGB2HSV)
plt.imshow(img_hsv2)
plt.show()

img_hsv3 = cv.cvtColor(img3, cv.COLOR_RGB2HSV)
plt.imshow(img_hsv3)
plt.show()
```

## i) Imprimir a matriz e o histograma da imagem HSV


```python
print("Imagem 1:\n")
print(img_hsv1)
plt.hist(img_hsv1.ravel(),256,[0,256])
plt.show()
print("Imagem 2:\n")
print(img_hsv2)
plt.hist(img_hsv2.ravel(),256,[0,256])
plt.show()
print("Imagem 3:\n")
print(img_hsv3)
plt.hist(img_hsv3.ravel(),256,[0,256])
plt.show()
```

## j) Transformar a imagem HSV em uma imagem com tons de cinza.

> Para isso, foi necessário transformar a imagem de HSV para RGB e, em seguida, para tons de cinza. Por isso, por já termos a imagem em RGB transformada em tons de cinza, as exibi novamente.


```python
print("Imagem 1:\n")
plt.imshow(img_gray1, cmap='gray')
plt.show()
print("\nImagem 2:\n")
plt.imshow(img_gray2, cmap='gray')
plt.show()
print("\nImagem 3:\n")
plt.imshow(img_gray3, cmap='gray')
plt.show()
```

## k) Imprimir a matriz e o histograma da imagem com tons de cinza


```python
print("Imagem 1:\n")
print(img_gray1)
plt.hist(img_gray1.ravel(),256,[0,256])
plt.show()
print("Imagem 2:\n")
print(img_gray2)
plt.hist(img_gray2.ravel(),256,[0,256])
plt.show()
print("Imagem 3:\n")
print(img_gray3)
plt.hist(img_gray3.ravel(),256,[0,256])
plt.show()
```

## l) Aplicar a inversa nas imagens nos tons de cinza


```python
img_neg1 = 255 - img_gray1
img_neg2 = 255 - img_gray2
img_neg3 = 255 - img_gray3

plt.imshow(img_neg1, cmap='gray')
plt.show()

plt.imshow(img_neg2, cmap='gray')
plt.show()

plt.imshow(img_neg3, cmap='gray')
plt.show()
```

## m) Aplicar o efeito de neblina nas imagens nos tons de cinza


```python
img_blur1 = cv.GaussianBlur(img_gray1, (5, 5), 0)
img_blur2 = cv.GaussianBlur(img_gray2, (5, 5), 0)
img_blur3 = cv.GaussianBlur(img_gray3, (5, 5), 0)

plt.imshow(img_blur1, cmap='gray')
plt.show()

plt.imshow(img_blur2, cmap='gray')
plt.show()

plt.imshow(img_blur3, cmap='gray')
plt.show()
```

## n) Aplicar a diminuição de contraste nas imagens nos tons de cinza


```python
img_gray_contrast1 = cv.convertScaleAbs(img_gray1, alpha=1.5, beta=0)
img_gray_contrast2 = cv.convertScaleAbs(img_gray2, alpha=1.5, beta=0)
img_gray_contrast3 = cv.convertScaleAbs(img_gray3, alpha=1.5, beta=0)

print("Imagem 1:\n")
plt.imshow(img_gray_contrast1, cmap='gray')
plt.show()
print("Imagem 2:\n")
plt.imshow(img_gray_contrast2, cmap='gray')
plt.show()
print("Imagem 3:\n")
plt.imshow(img_gray_contrast3, cmap='gray')
plt.show()
```
